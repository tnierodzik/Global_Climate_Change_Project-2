```python
#Import os and csv modules
import os
import csv
import pandas as pd
```

```python
# File to Load (Remember to Change These)
co2_emmission = "Data Resources/Emissions/annual-co-emissions-by-region.csv"
lat_lng = "Data Resources/Countries Lat & lng.xlsx"
energy_consumption = "Data Resources/Energy Consumption/global-primary-energy.csv"
```

```python
co2_emission_df = pd.read_csv(co2_emmission)
co2_emission_df
```

```python
lat_lng_country = pd.read_excel(lat_lng)
lat_lng_country = lat_lng_country.rename(columns = {"country" : "Code", "name" : "Entity"})
lat_lng_country
```

```python
new_df = co2_emission_df.merge(lat_lng_country, on = "Entity", how = "outer")
new_df
```

```python
new_df.dropna()
```

```python
new_df.to_csv("co2_emission.csv")
```
