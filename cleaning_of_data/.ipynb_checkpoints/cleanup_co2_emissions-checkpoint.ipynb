```python
# Read in world json data
with open('../Resources/Coordinate Data/minified_world_data.json', 'r') as f:
    json_data = f.read()
    
json_d = json.loads(json_data)
```

```python

# Read in temp data
temp_data_file = '../Resources/Global Temperature Changes by Country/GlobalLandTemperaturesByCountry.csv'
temp_data_df = pd.read_csv(temp_data_file, parse_dates=True)
new_temp = temp_data_df.dropna()
new_temp
```

```python

# Read in emissions data
emissions_data_file = '../Resources/Emissions/annual-co-emissions-by-region.csv'
emissions_data_df = pd.read_csv(emissions_data_file)
emissions_data_df.head()
```

```python
# Cleaning temp data - calculate yearly temp average per year (1950-2013)
new_temp['dt']=pd.to_datetime(new_temp['dt'])
new_temp['year'] = pd.PeriodIndex(new_temp.dt, freq='Y')
new_temp['year'].value_counts().sort_index() #Do i need sort_index()?

yearly_average_temp = new_temp.groupby(['Country', 'year'])['AverageTemperature'].mean()
yearly_average_temp

# yearly_average_temp_df = yearly_average_temp
# yearly_average_temp_df =
```

```python

for country in enumerate(json_d['features']):
    for i in range(1950, 2013):
        country['properties'][f'year_temp{i}'] = yearly_average_temp[country['properties']['name'], i]
        country['properties'][f'year_emission{i}'] = emissions_data_df[country['properties']['name'], i]
```
